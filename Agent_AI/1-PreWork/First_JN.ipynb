```python
fruit = "apple"
print(fruit)

letter = fruit[0]
print(letter)
```

```python
len(fruit)          # Get the length of the string

fruit.lower()       # Convert to lowercase
fruit.upper()       # Convert to uppercase
fruit.capitalize()  # Capitalize the first letter

"    Hello, World!    ".strip()  # Remove leading and trailing whitespace
```

```python
"a, b, c".split(", ")  # Split the string into a list
"-".join(["a", "b", "c"])  # Join a list into a string with a separator
"hello world".replace("o", "0")  # Replace characters in the string
"hello world".find("o")  # Find the index of the first occurrence of a character
```
