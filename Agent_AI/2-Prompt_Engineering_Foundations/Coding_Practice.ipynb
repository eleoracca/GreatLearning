Question 2

You are given a dictionary as input. Write a function to find the key with the maximum value in the dictionary and return it.

Test Cases:
1. {'a': 10, 'b': 20, 'c': 5} 
   expected output: b
2. {'x': 100, 'y': 50, 'z': 75}
   expected output: x

```python
def find_max_key(my_dict):
    # Write your code below
    max_key = None
    max_value = float("-inf")

    for key, value in my_dict.items():
        if value > max_value:
            max_value = value
            max_key = key

    return max_key

my_dict = eval(input())
print(find_max_key(my_dict))
```

You are given an array nums containing n distinct integers taken from the range 0 to n, inclusive. However, one of the integer is missing from the array, and your task is to find and return this missing integer.
To clarify, the array nums contains integers from 0 to n, with no duplicates, and exactly one integer missing. Your goal is to implement a functino that takes the input array nums and returns the missing integer.

Hints:
- Use the formula n*(n+1)/2 to find the sum of the first n natural numbers.
- Subtract the sum of the given array from the sum of the first n natural numbers to find the missing number.
- You can use the sum() function to find the sum of the given array.

Test Cases:
1. [3, 0, 1]
   expected output: 2
2. [0, 1, 2, 4, 5]
   expected output: 3

```python
def missing_number(nums):
    # Write your code below
    n = len(nums)
    sum_expected = n * (n+1) // 2
    sum_actual = sum(nums)

    return sum_expected-sum_actual

nums = eval(input())
print(missing_number(nums))
```

Question 7

You are given an integer as input. Write a program to find the prime factors of the given number and return them as a list.

Hints:
- Start with the smallest prime number, 2, and check if it divides the given number evenly.
- If it does, add 2 to the list of prime factors and divide the given number by 2.
- Repeat the process with the next prime number until the given number becomes 1.

Test Cases:
1. 60
   expected output: [2, 2, 3, 5]
2. 24
   expected output: [2, 2, 2, 3]

```python
def prime_factorization(n):
    factors = []

    while n % 2 == 0:
        factors.append(2)
        n //= 2

    divisor = 3
    while divisor * divisor <= n:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 2

    if n > 1:
        factors.append(n)

    return factors

n = int(input())
print(prime_factorization(n))
```

You are given an unordered list with consecutive integers starting from 1. Write a program to find the minimum number of swaps required to sort the list in ascending order.

Hints:
- You can use a dictionary to keep track of the indices of the elements in the list.
- Iterate through the list and check if the current element is in the correct position. If not, swap it with the element at the correct position and increment the swap count.
- Repeat step 2 until the list is sorted.

Test Cases:
1. [7, 1, 3, 2, 4, 5, 6]
   expected output: 5
2. [3, 2, 1]
   expected output: 1

```python
input_list = eval(input())

def min_swaps(nums):
    index_map = {value: i for i, value in enumerate(nums)}
    swaps = 0
    n = len(nums)

    for i in range(n):
        correct_value = i + 1

        if nums[i] != correct_value:
            swaps += 1

            to_swap_idx = index_map[correct_value]

            index_map[nums[i]] = to_swap_idx
            index_map[correct_value] = i

            nums[i], nums[to_swap_idx] = nums[to_swap_idx], nums[i]

    return swaps

print(min_swaps(input_list))
```

You are given a collection of intervals, where each interval is represented as a list [start, end], denoting the start and end points of the interval, respectively. Your task is to merge all overlapping intervals in the collection.
To clarify, if any two intervals in the collection overlap or share any common points, they should be merged into a single interval. The merged interval should span the combined range of the original overlapping intervals.
Your objective is to implement a function that takes the input list of intervals and returns a new list containing the merged intervals.

Hints:
- Sort the intervals based on the start point.
- Initialize an empty list to store the merged intervals.
- Iterate through the sorted intervals and check if the current interval overlaps with the previous merged interval. If it does, update the end point of the mergec interval. If it doesn't, add the current interval to the merged intervals list.

Test Cases:
1. [[1, 3], [2, 6], [8, 10], [15, 18]]
   expected output: [[1, 6], [8, 10], [15, 18]]
2. [[1, 4], [4, 5]]
   expected output: [[1, 5]]

```python
def merge_intervals(intervals):
    if not intervals:
        return []

    intervals.sort(key=lambda x: x[0])

    merged = [intervals[0]]

    for current in intervals[1:]:
        last = merged[-1]

        if current[0] <= last[1]:  # overlap
            last[1] = max(last[1], current[1])
        else:
            merged.append(current)

    return merged


intervals = eval(input())
print(merge_intervals(intervals))
```

Question 10

You are given a string consisting of only the characters '(', ')', '{', '}', '[', ']'. Your task is to determine whether the input string is valid. An input string is considered valid if:
- Open brackets must be closed by the same type of brackets.
- Open brackets must be closed in the correct order.
- An empty string is also considered valid.

Hints:
- Use a stack data structure to keep track of the opening brackets.
- Iterate through the characters of the string and check if each character is an opening or a closing bracket.
- If it is an opening bracket, push it onto the stack.
- If it is a closing bracket, check if the stack is empty or if the top of the stack is the corresponding opening bracket.
- If the stack is empty or the top of the stack is not the corresponding opening bracket, the string is not valid.
- After iterating through all the characters, check if the stack is empty. If it is empty, the string is valid.

Test Cases:
1. ()[]{}
   expected output: True
2. ([)]
   expected output: False

```python
def is_valid(s):
    stack = []
    
    pairs = {
        ')': '(',
        ']': '[',
        '}': '{'
    }

    for char in s:
        if char in "([{":
            stack.append(char)
        else:
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()

    return len(stack) == 0

s = input().strip()
print(is_valid(s))
```
